```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
```

```python
def plot_data(sample):
    plt.figure(figsize=(10,8))
    plt.subplot(1,2,1)
    sns.histplot(sample,kde=True)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist="norm",plot=pylab)
    plt.show()
```

```python
plot_data(np.log(s2))
```

```python
plot_data(s2)
```

```python
plot_data(df["PetalLengthCm"])
```
