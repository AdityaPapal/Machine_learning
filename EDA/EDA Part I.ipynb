```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.suptitle("Univariate Analysis of Catgorical Features",fontsize = 20,fontweight='bold',alpha=0.8,y=1)
catgory = ['Content Rating']
for i in range(0,len(catgory)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df[catgory[i]],palette="Set2")
    plt.xlabel(catgory[i])
    plt.xticks(rotation = 45)
    plt.tight_layout()
```
