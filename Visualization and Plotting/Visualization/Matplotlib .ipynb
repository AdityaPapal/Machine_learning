```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

## Scatter Plot

```python
plt.scatter(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("2D Graph")
```

## Plot

```python
plt.plot(x,y,'r')
plt.xlabel("X-Axies")
plt.ylabel("Y-Axies")
plt.title("Graph of plot")
```

## Subplot

```python
# subplot
y = x*x
plt.subplot(3,2,1)
plt.plot(x,y,'r')

plt.subplot(3,2,2)
plt.plot(x,y,'g')

plt.subplot(3,2,3)
plt.plot(x,y,'m')

plt.subplot(3,2,4)
plt.plot(x,y,'y')

plt.subplot(3,2,5)
plt.plot(x,y,'k')

plt.subplot(3,2,6)
plt.plot(x,y,'c')
```

## Histogram

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
```

```python
plt.hist(a,bins = 20) 
plt.title("histogram") 
```

## Box Plot

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);  
```

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
